```python
import numpy as np
```

 Graficaremos la funcio Seno dandole valores. Lo primero
 que haremos es definir la funcion 
Le colocamos un nombre cualquiera m y le pasamos una variable x, luego con la palabra reservada return 
nos retornara la funcion seno, np.sin(x),Luego tenemos que asignarle los valores para saber desde donde hasta donde me graficara
la funcion seno con linspace nos da una lista de valores que puede tomar x
Donde Comienza,Donde Termina, y el Numero de puntos que queremos en ese rango
Ejemplo de 0 a 10 y 100 puntos en ese rango.
Luego con plt.plot graficamos pasandole como argumentos
variable que depende la funcion que vamos a graficar en este caso x Luego la funcion a graficar m(x)

```python
def m(x):
    return np.sin(x)
x=np.linspace(0,10,100)
plt.plot(x,m(x))
```

Colocamos el nombre a los ejes y - x

```python
plt.xlabel('Tiempo')
```

```python
plt.ylabel('posicion')
```

```python
plt.title('FUNSION SENO')
```
